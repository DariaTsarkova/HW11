```python
import pandas as pd

movies = pd.read_csv(r'C:\Users\Dasha\HW11\movies.csv') # загружаем файл
ratings = pd.read_csv(r'C:\Users\Dasha\HW11\ratings.csv') # загружаем файл
joined = movies.merge(ratings, on='movieId', how='left') # объединяю два DataFrame по столбцу movieId, используя левое соединениеheads
joined.head()
```

```python
def classyfy(rating):
    if 0 < rating <= 2:
        return 'низкий рейтинг'
    elif rating <= 4:
        return 'средний рейтинг'
    elif 4.5 <= rating <= 5:
        return 'высокий рейтинг'
    else:
        return 'ошибка в данных'
joined['class'] = joined['rating'].apply(classyfy)
joined.head(50)
```
